```python
import random

positive_words = {
    "good", "great", "excellent", "amazing", "awesome",
    "love", "happy", "best", "fantastic", "wonderful",
    "nice", "super", "perfect", "outstanding",
    "brilliant", "enjoy", "liked", "beautiful"
}

negative_words = {
    "bad", "worst", "hate", "terrible", "awful",
    "poor", "sad", "disappointed", "horrible",
    "useless", "boring", "problem", "issue",
    "annoying", "pathetic", "dislike", "ugly"
}

positive_emojis = [
    "😊", "😀", "😁", "😍", "🥰", "😎",
    "👍", "❤️", "🎉", "✨", "💖", "😄"
]

negative_emojis = [
    "😢", "😭", "😡", "😠", "👎",
    "💔", "😞", "😔", "🤬", "😣"
]


def analyze_sentiment(text):
    words = text.lower().split()
    positive_count = 0
    negative_count = 0

    for word in words:
        word = word.strip(".,!?()[]{}\"'")
        if word in positive_words:
            positive_count += 1
        elif word in negative_words:
            negative_count += 1

    for char in text:
        if char in positive_emojis:
            positive_count += 1
        elif char in negative_emojis:
            negative_count += 1

    total = positive_count + negative_count

    if total == 0:
        sentiment = "Neutral"
        confidence = 0.0
    elif positive_count > negative_count:
        sentiment = "Positive"
        confidence = (positive_count / total) * 100
    elif negative_count > positive_count:
        sentiment = "Negative"
        confidence = (negative_count / total) * 100
    else:
        sentiment = "Neutral"
        confidence = 50.0

    if confidence >= 90:
        confidence_label = "Very High"
    elif confidence >= 70:
        confidence_label = "High"
    elif confidence >= 50:
        confidence_label = "Medium"
    else:
        confidence_label = "Low"

    return sentiment, confidence, confidence_label, positive_count, negative_count


print("=" * 50)
print("ADVANCED SENTIMENT ANALYZER 😊🚀")
print("=" * 50)

while True:
    review = input("\nEnter Review (type 'exit' to quit): ")

    if review.lower() == "exit":
        print("\nProgram Closed Successfully! 👋")
        break

    sentiment, confidence, label, pos, neg = analyze_sentiment(review)

    # Random emoji selection for display
    pos_emoji = random.choice(positive_emojis)
    neg_emoji = random.choice(negative_emojis)

    print("\n----------- RESULT -----------")
    print("Review           :", review)
    print("Sentiment        :", sentiment)
    print(f"Confidence       : {confidence:.2f}%")
    print("Confidence Label :", label)
    print("Positive Score   :", pos, pos_emoji)
    print("Negative Score   :", neg, neg_emoji)
    print("------------------------------")
```
